```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
date_range = pd.date_range(start='2025-01-01', periods=15, freq='D')
temperature = [10, 5, 15, 8, None, 13, None, 11, 12, 9, None, 7, 6, 14, None]
weather_conditions = ['Sunny', 'Rainy', 'Cloudy', 'Sunny', None, 'Cloudy', 'Rainy', 'Sunny', 'Cloudy', 'Cloudy', None, 'Sunny', 'Rainy', 'Sunny', None]
```

```python
df = pd.DataFrame({
    'Date': date_range,
    'Temperature (°C)': temperature,
    'Weather Condition': weather_conditions
})
```

```python

print("Original Dataset with Missing Values:")
print(df)
```

```python
df_original = df.copy()
```

```python
plt.figure(figsize=(12, 6))

# Plot Temperature (°C) Before Interpolation
plt.subplot(1, 2, 1)
plt.plot(df_original['Date'], df_original['Temperature (°C)'], marker='o', label='Original', color='red', linestyle='dashed')
plt.title('Temperature Before Interpolation')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
```

```python
df['Temperature (°C)'] = df['Temperature (°C)'].interpolate(method='linear')
```

```python
df['Weather Condition'] = df['Weather Condition'].fillna(method='ffill')
```

```python
print("\nDataset After Interpolation:")
print(df)
```

```python
plt.subplot(1, 2, 2)
plt.plot(df['Date'], df['Temperature (°C)'], marker='o', label='Interpolated', color='red')
plt.title('Temperature After Interpolation')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
